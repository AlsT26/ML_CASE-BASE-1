###Menghitung Outlier

```python
def hitungOutliers(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    IQR = q3 - q1
    outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
    return outliers
```

```python
kurang5persen = []
diatas10persen = []
for i in df:
  outliers = hitungOutliers(df[i])
  percentage = len(outliers)/len(df[i])*100
  if percentage <= 3:
    kurang5persen.append(i)
  if percentage > 10:
    diatas10persen.append(i)
print("Kolom dengan outlier kurang dari 5% : ",kurang5persen)
print("Kolom dengan outlier diatas 10% : ",diatas10persen)
```

```python
def mepetOutliers(df):
  for i in kurang5persen:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)

    IQR = Q3 - Q1

    LB = Q1 - (IQR * 1.5)
    UB = Q3 + (IQR * 1.5)

    df.loc[df[i] > UB, i] = UB
    df.loc[df[i] < LB, i] = LB
df_mepetOutliers = df
mepetOutliers(df_mepetOutliers)
```

#ANN Model
Disini library yang digunakan adalah Tensorflow

```python
import tensorflow as tf
from tensorflow.keras.optimizers import RMSprop
```

```python
Create_Model =  tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(32, activation='leaky_relu'),
    tf.keras.layers.Dropout(0.2), 
    tf.keras.layers.Dense(17, activation='softmax')
])
```

```python
Create_Model.compile(optimizer=tf.keras.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
Create_Model.fit(X_train,y_train,epochs=500,batch_size=20)
```

```python
Create_Model.evaluate(X_train,y_train)
```

```python
predictions = Create_Model.predict(X_test)
```

```python
a = 0
b = 0

for prediction, y in zip(predictions, y_test):
  if np.argmax(prediction) == y:
    a = a + 1
  else :
    b += 1
  print(f"Prediction: {np.argmax(prediction)}, Hasil Y_Test :{y}")
print("Akurasi : ",(a/(a+b) * 100))
```
